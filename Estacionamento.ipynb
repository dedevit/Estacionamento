```python
from typing_extensions import ParamSpecArgs

class Estacionamento:
  """
    self.__vagas_de_carro e self.__vagas_de_moto - são respectivamente os idvagas. do tipo lista
    self.__total_vagas_livres_carro e self.__total_vagas_livres_moto    inicialisa respectivamente o numero de vagas para motos e carros
    self.__tipo_carro  e self.__tipo_moto  - são codigos para os tipos. Optei por numeros visto que nomes teriamos que fazer verificações para Maiuscula e miniscula e outros possiveis erros
    self.__vagas_carro_ocupada  eself.__vagas_moto_ocupada   - inicialisa a lista com as vagas que serão ocupadas
    self.____capacidade_maxima  - criei esse atributo para o caso de precisar aumentar o numero de vagas, os calculos feitos dentro de estado do estacionamento serão atualizados automaticamente
  
  """
  def __init__(self):
    self.__vagas_de_carro = [2,4,6,8,10]
    self.__vagas_de_moto = [111,113,115,117,119]
    self.__total_vagas_livres_carro = 5
    self.__total_vagas_livres_moto = 5
    self.__tipo_carro = 0
    self.__tipo_moto = 11
    self.__vagas_carro_ocupada  = []
    self.__vagas_moto_ocupada  = []
    self.__capacidade_maxima = 10
  
   

  def estacionar_carro(self, tipo, placa, idvaga):
    """
     Método que  verifica se há disponibilidade de vagas, estaciona o carro ou moto e retorna mensagem e realiza o calculo de acordco com o tipo de veiculo.
    """
   
    if not( tipo == 0 or tipo == 11):
      return print(f'Tipo inválido. Escolha {self.__tipo_carro} para carro ou {self.__tipo_moto } moto')


    if not self.__total_vagas_livres_carro  >= len(self.__vagas_de_carro) : # se o valor for menor que o numero de vagas disponíveis imprime mensagem.
      
      return print('Todas as vagas estão ocupadas. Volte mais tarde')
    
    if not (idvaga in  self.__vagas_de_carro or idvaga in  self.__vagas_de_moto):
      return f'Identificador da vaga não existe.Por favor escolha outra vaga'
       
    if idvaga in self.__vagas_carro_ocupada : # verifica se a vaga está disponível
      return print('Vaga ocupada.Escolha outra vaga')

    if (tipo ==11): # verifica se o tipo do veiculo é moto ou carro e realiza calculos e retorna mensagem de acordo com o tipo definido
      self.__vagas_carro_ocupada .append(idvaga)
      self.__vagas_de_carro.remove(idvaga)
      return f' Moto placa {placa}, está estacionado na vaga  de carro nº {self.__vagas_carro_ocupada }'

    else:
      self.__vagas_carro_ocupada .append(idvaga)
      self.__vagas_de_carro.remove(idvaga)
      return f' Carro placa {placa}, está estacionado na vaga nº  {self.__vagas_carro_ocupada }'


  def estacionar_moto(self, tipo, placa, idvaga): 
    """
     Método que  verifica se há disponibilidade de vagas,se o veiculo é moto pelo tipo e  retorna mensagem 
    """
    
    if not self.__total_vagas_livres_moto  >= len(self.__vagas_de_moto) and self.__total_vagas_livres_carro  >= len(self.__vagas_de_carro)  :

      return print('Todas as vagas estão ocupadas. Volte mais tarde')
      
    
    if tipo != 11:
     return print(f'Tipo inválido, vagas somente para motos. Use o tipo {  self.__tipo_moto}. ')
    
    if not idvaga in  self.__vagas_de_moto:
       return f'Identificador da vaga não existe.Por favor escolha outra vaga. Vagas de moto {self.__vagas_de_moto}'
      
    
    if idvaga in self.__vagas_moto_ocupada :
      return print('Escolha outra vaga')

    else:
     self.__vagas_moto_ocupada .append(idvaga)
     self.__vagas_de_moto.remove(idvaga)
     
     return f'Moto placa {placa}, vaga ocupada {self.__vagas_moto_ocupada }'


  def estado_estacionamento(self):
    """
    Método que verifica a ocupação do estacionamento usando os atributos  vagas carro e moto ocupados e vagas livres carro e moto todos inicializados com valor igual a 5
    e operados dentro dos metodos estacionar carro e moto removendo e adicionando vagas.
    """
    total_vagas_ocupadas = len(self.__vagas_carro_ocupada ) +  len(self.__vagas_moto_ocupada ) 

    if  total_vagas_ocupadas  == 10:
      print('Todas as vagas destinadas a motos e carros estão ocupadas')

    else:
      total_vagas_disponíveis_carro =self.__total_vagas_livres_carro -  len(self.__vagas_carro_ocupada )
      total_vagas_disponíveis_moto = self.__total_vagas_livres_moto  -   len(self.__vagas_moto_ocupada )
      total_vagas_disponiveis =  self.__capacidade_maxima = 10 - (len(self.__vagas_carro_ocupada ) + len(self.__vagas_moto_ocupada ) )

    return f' Numero total de vagas disponíveis: {total_vagas_disponiveis}.Numero total de vagas disponíveis para carros: { total_vagas_disponíveis_carro}, vagas: {self.__vagas_de_carro}. Numero total de vagas disponíveis para motos: { total_vagas_disponíveis_moto}, vagas: { self.__vagas_de_moto}. '
  
  def remover_carro(self,idvaga):

    """
    Método que libera a vaga. Utiliza os atributos vagas carro ocupada e vagas carro. Verifica se o numero da vaga digitada existe e retorna uma mensagem.
    """

    if not idvaga in self.__vagas_carro_ocupada :

       return f'Vaga não existe ou não está ocupada. Vagas de carros disponíveis {  self.__vagas_de_carro}'

    self.__vagas_carro_ocupada .remove(idvaga)
    self.__vagas_de_carro.append(idvaga)
      
    return f'A vaga {idvaga} foi liberada com suceso. Vagas de carros disponíveis {  self.__vagas_de_carro}'

  
  def remover_moto(self,idvaga):

    """
    Método que libera a vaga. Utiliza os atributos vagas moto ocupada e vagas moto. Verifica se o numero da vaga digitada existe e retorna uma mensagem.
    """

    if not idvaga in self.__vagas_moto_ocupada :
       return f'Vaga não existe ou não está ocupada. Vagas de motos disponíveis {  self.__vagas_de_moto}'

    self.__vagas_moto_ocupada .remove(idvaga)
    self.__vagas_de_moto.append(idvaga)
      
    return f'A vaga {idvaga} foi liberada com suceso. Vagas de moto disponíveis {  self.__vagas_de_moto}'
      
      

      




    
   


  

       

```

```python
#Exemplo - Encapsulamento
# Getter = retornar o valor 
# Setter = definir ou atualizar

 
class Quadrado:
  def __init__(self):
    self.__lado = 1
  

  @property # getter
  def lado(self):
    return self.__lado

 

  @lado.setter #setter
  def lado(self, valor):
    if valor < 0:
      raise ValueError()
    self.__lado = valor

  def area(self):
    area= self.lado * self.lado
    return f' A  area do quadrado de lado igual a L= { self.__lado} é  Área ={area}'

 

quadrado = Quadrado()
quadrado.lado=2
print(quadrado.area())


```

```python
#HERANÇA.
#eXEMPLO FUNCIONARIOS. lINK: https://jacksongomesbr.gitbooks.io/programacao-orientada-a-objetos-com-python/content/heranca.html

class Pessoa:
  def __init__(self, nome):
    self._nome = nome
    self._tipo = 'Pessoa'

 

  def falar_oi(self):
    print('Oi, meu nome é {}'.format(self._nome))

 

  def falar_tipo(self):
    print('Meu tipo é {}'.format(self._tipo))

 

pessoa = Pessoa('Maria')
pessoa.falar_oi()
pessoa.falar_tipo()

 

# A classe estudante é derivada da classe Pessoa
# Relação: Estudante é uma pessoa
class Estudante(Pessoa): 
  def __init__(self, nome, curso):
    super().__init__(nome) #chama o construtor na classe base
    self._curso = curso

 

  def falar_curso(self):
    print(f'Eu, {self._nome}, estudo o curso {self._curso}') #A propriedade self._nome é herdada da classe base

 

  def falar_tipo(self): 
    self._tipo = 'Estudante'
    return super().falar_tipo()

 

estudante = Estudante('Yasmin', 'Introdução ao Python')
estudante.falar_oi() # o método "falar_oi" é herdado da classe base
estudante.falar_tipo() #o método "falar_tipo" é herdado da classe base, e sobreescrito na classe derivada
estudante.falar_curso
```

```python
#HERANÇA MULTIPLA

class Logavel:
  def __init__(self):
    self.nome_da_classe = ''
  def logar(self, mensagem):
    print('Mensagem da classe' + self.nome_da_classe + ':'+mensagem)

 

class Conexao:
  def __init__(self):
    self.servidor = ''
  def conectar(self):
    print('Conectando ao banco de dados no servidor' + self.servidor)
    # Lógica para realizar a conexão com o BD

 

class MySqlDatabase(Conexao, Logavel):
  def __init__(self):
    super().__init__()
    self.nome_da_classe = 'MySqlDataBase'
    self.servidor = 'Meu servidor'

 

def framework(item):
  if isinstance(item, Conexao):
    item.conectar()
  if isinstance(item, Logavel):
    mensagem = 'Boa noite, minhas queridas. SABADOUUUU'
    item.logar(mensagem)
    
conexao_mysql = MySqlDatabase()
framework(conexao_mysql)
```
